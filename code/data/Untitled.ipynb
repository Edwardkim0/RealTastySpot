```python
import pandas as pd
import numpy as np
import csv
```

```python
num1 = pd.read_csv('2020-09-01_15_44_linkkakao_review.csv', sep='\t').drop('Unnamed: 0', axis=1)
num2 = pd.read_csv('2020-09-01_16_24_linkkakao_review.csv', sep='\t').drop('Unnamed: 0', axis=1)
```

```python
total_kakao = pd.concat([num1, num2])
```

```python
total_kakao['num_review'] = total_kakao.scores.map(lambda x:len(x.split(',')))
```

```python
total_kakao = total_kakao.drop('length', axis=1)
```

```python
total_kakao.name = total_kakao.name.map(lambda x:'경천애인' if '경천' in x else x)
```

```python
total_kakao.name = total_kakao.name.map(lambda x:'나폴레옹' if '나폴레옹' in x else x)
```

```python
total_kakao.to_excel('kakao_review.xlsx', index=False)
```
